```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import LinearRegression, TheilSenRegressor,RANSACRegressor,HuberRegressor
from sklearn.metrics import r2_score
from xgboost import XGBRegressor
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
test.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectFromModel
```

```python
train_labels = train.iloc[:,1]
train_feats = train.iloc[:, 2:]

test_labels = test.iloc[:,1]
test_feats = test.iloc[:, 1:]
```

```python
train_feats_encode = []
train_feats_encode.append(train.iloc[:, 0])
for i in range(train_feats.shape[1]):
    arr = train_feats.iloc[:, i]
    if arr.dtype == 'O':        
        lblencod = LabelEncoder()
        arr = lblencod.fit_transform(arr)
    train_feats_encode.append(arr)
train_feats_encode = np.array(train_feats_encode).T

test_feats_encode = []
test_feats_encode.append(test.iloc[:, 0])
for i in range(test_feats.shape[1]):
    arr = test_feats.iloc[:, i]
    if arr.dtype == 'O':        
        lblencod = LabelEncoder()
        arr = lblencod.fit_transform(arr)
    test_feats_encode.append(arr)
test_feats_encode = np.array(test_feats_encode).T
```

```python
cellen=['X47','X95','X314','X315','X232','X119','X311','X76','X329','X238','X340','X362','X137', 'ID']
train_feats_encode = train[cellen]
test_feats_encode = test[cellen]
```

```python
model_xgb2 = XGBRegressor(seed = 0,
      colsample_bytree = 0.7,
      subsample = 0.9,
      eta = 0.03,
      max_depth = 4,
      num_parallel_tree = 1,
      min_child_weight = 1, objective='reg:linear', base_score=np.mean(train_labels))
```

```python
def check(model, i=1):
    total = 0.
    for _ in range(0, i):
        t_X, q_X, t_y, q_y = train_test_split(train_feats_encode, train_labels)
        model.fit(t_X, t_y)
        print(r2_score(t_y, model.predict(t_X)))
        s = r2_score(q_y, model.predict(q_X))
        print(s)

        model.fit(q_X, q_y)
        print(r2_score(q_y, model.predict(q_X)))
        s2 = r2_score(t_y, model.predict(t_X))
        print(s2)
        avg = 1.* (s + s2) /2 
        total += avg
        print("average : ", avg)
        print("==============================")
    print("\nTotal Avg : ", total * 1. /i)
```

```python
check(model_xgb2, 5)
```

```python
model_xgb2.fit(train_feats_encode, train_labels)
print(r2_score(train_labels, model_xgb2.predict(train_feats_encode)))
```

```python
output = pd.DataFrame({'id': test['ID'].astype(np.int32), 'y': model_xgb2.predict(test_feats_encode)})
output.to_csv('results/xgboost/ensemble[%s][%s].csv' % ('ensemble',avg), index=False)
```
